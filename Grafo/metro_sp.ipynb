Definição dos nós e arestas

```python
G = nx.Graph()

for idx, row in df[['nome0', 'nome1', 'cor0', 'cor1']].iterrows():

    # Adicionando os nós e arestas
    G.add_edge(row[0], row[1])
    # Atualizando as cores
    G.add_node(row[0], color=row[2])
    G.add_node(row[1], color=row[3])
```

Visualização do grafo

```python
# Obtém as posições dos nós
pos = nx.kamada_kawai_layout(G, scale=5.0)

# Desenha o grafo
plt.figure(figsize=(30,30))
colors = nx.get_node_attributes(G, 'color')
nx.draw(G, pos, node_color=[colors[v] for v in G], with_labels=True, font_size=10)
plt.show()
```
