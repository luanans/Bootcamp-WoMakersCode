```python
import datetime
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

# Aula 1

```python
df = pd.read_csv('monitoramento_tempo.csv')
df.head()
```

```python
df.info()
```

```python
df.data = pd.to_datetime(df.data)
```

```python
plt.figure(figsize=(18,6))
plt.plot(df.data, df.temperatura)
plt.title('Temperatura X Tempo', fontsize=16)
```

```python
fig = plt.figure(figsize=(18,6))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df.data, df.temperatura, color='slateblue')
eixo.set_title('Temperatura X Tempo', fontsize=16)
eixo.set_ylabel('Temperatura (K)', fontsize=14)
eixo.set_xlabel('Anos', fontsize=14)
eixo.legend(['Temperatura'], loc='lower right', fontsize=14)
plt.show()
```
