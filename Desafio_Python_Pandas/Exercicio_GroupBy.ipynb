#### Como usar o groupby()

```python
from pydataset import data
import pandas as pd
```

```python
# Importação dos dados
df = data('Forbes2000')
df.head(2)
```

```python
# Um grupo e uma coluna
df.groupby(['category'])['profits'].sum()
```

```python
# Multiplas colunas
df.groupby(['category'])[['assets', 'sales', 'profits']].sum()
```

```python
# Multiplos grupos
df.groupby(['category'], as_index=False)[['assets', 'sales', 'profits']].sum()
```

```python
# Grupos como colunas
df.groupby(['category', 'country'], as_index=False)[['assets', 'sales', 'profits']].sum()
```

```python
# Multiplas funções
df.groupby('category')[['assets', 'sales']].agg(['mean', 'sum'])
```
